### 1:  Create some data

```python
import numpy as np
X = np.array([[1,2],[2,2],[3,1],[5,-0.5],[1,1],[2,0],[0.5,1],[-1,2]])
y = np.array([1,1,1,1,0,0,0,0])
```

### 2: Class

```python
class LogisticRegression():
    def __init__(self,X,y,lr,init_method,epoch,scope):
        self.X = X
        self.y = y
        self.lr = lr
        self.scope=scope
        self.init_method = init_method
        self.wb = np.zeros((1,self.X.shape[1]+1))
        self.epoch = epoch
    
    def get_init_wb(self):
        if self.init_method=='zero':
            return np.zeros((1,self.X.shape[1]+1))
        elif self.init_method=='rand':
            return np.random.rand(1, self.X.shape[1]+1)
        elif self.init_method=='randint':
            return np.random.randint(0, 10, size=(1, self.X.shape[1]+1))
        else:
            return init_method
        
    def update_wb(self, X, y):
        delta_w = y * X-np.dot((np.exp(np.dot(self.wb[0,0:-1], X.T)+self.wb[0,-1])/   \
                           (1 + np.exp(np.dot(self.wb[0,0:-1], X.T)+self.wb[0,-1]))), X)
        delta_b = y - np.exp(np.dot(self.wb[0,0:-1], X.T)+self.wb[0,-1])/   \
                           (1 + np.exp(np.dot(self.wb[0,0:-1], X.T)+self.wb[0,-1]))
        self.wb[0,0:-1] = self.wb[0,0:-1] + self.lr*delta_w
        self.wb[0, -1] = self.wb[0, -1] + self.lr*delta_b
        #print(self.wb)
        
    def calculate_acc(self):
        com_cou = np.exp(np.dot(self.wb[0,0:-1], X.T)+self.wb[0,-1])/   \
                           (1 + np.exp(np.dot(self.wb[0,0:-1], X.T)+self.wb[0,-1]))
        return com_cou
    
    def train(self):
        self.wb = self.get_init_wb()
        for i in range(self.epoch):
            for j in range(self.X.shape[0]):
                self.update_wb(self.X[j], y[j])
            if i%self.scope==0:
                print('epoch {} '.format(i+1),'accuracy: ',self.calculate_acc())
        print('epoch {} '.format(i+1),'accuracy: ',self.calculate_acc())
        print('-------------training finished!!!-------------')
    
    def predict_prob(self, data):
        com_cou = np.exp(np.dot(self.wb[0,0:-1], data.T)+self.wb[0,-1])/   \
                    (1 + np.exp(np.dot(self.wb[0,0:-1], data.T)+self.wb[0,-1]))
        return com_cou
```

```python
mylr = LogisticRegression(X,y,0.5,'zero',40, 1)
mylr.train()
```
